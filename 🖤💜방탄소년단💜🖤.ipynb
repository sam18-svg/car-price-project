```python
df=df.drop(columns={"make","model","market_category"},axis=1)
```

```python
df.head()
```

```python
df['engine_hp'].isnull()
```

```python
df['engine_hp']=df['engine_hp'].fillna (df['engine_hp'].mean())
```

```python
df['engine_cylinders']=df['engine_cylinders'].fillna (df['engine_cylinders'].mean())
```

```python
df['number_of_doors']=df['number_of_doors'].fillna (df['number_of_doors'].mean())
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df=pd.get_dummies(df , columns=['engine_fuel_type' ,'vehicle_size','driven_wheels','vehicle_style','transmission_type'])
```

```python
df.head()
```

```python
x=df.drop(columns='price',axis=1)
y=df['price']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from numpy.random.mtrand import random
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error ,r2_score
```

```python
error=mean_squared_error(y_test,prediction)
```

```python
print(error)
print(np.sqrt(error))
```

```python
r2_score(y_test,prediction)
```

```python
import joblib as jb
```

```python
jb.dump(model,"car_price_model1.pkl")
```
